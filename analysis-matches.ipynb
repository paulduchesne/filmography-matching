```python
# highest maximum match per agent.

import altair
import numpy
import pandas
import pathlib
import shutil

altair.data_transformers.disable_max_rows()

median_path = pathlib.Path.cwd() / 'median'
median_data = [x for x in median_path.iterdir() if x.suffix == '.csv']

print(len(median_data), 'csvs.')

analysis1 = pandas.DataFrame()

for x in median_data:
    dataframe = pandas.read_csv(x)
    analysis1 = pandas.concat([dataframe[:1], analysis1])

altair.Chart(analysis1).mark_bar().encode(
    altair.X('match:N',  title='maximum matching score'), 
    altair.Y('count(acmi_id):Q',  title='count of occurances')
).properties(
    width=1000, 
    height=400)
```

```python
for x in range(80,100)[::-1]:
    examples = analysis1.copy()
    examples = examples.loc[examples.match.isin([x])][:4]
    print('***', x, 'match')
    print('\n')
    for y in examples.to_dict('records'):
        state = f"{y['acmi_name'].title()} [{y['acmi_id']}] same as "
        state += f"{y['wikidata_name'].title()} [{y['wikidata_id']}]."
        print(state)
        data = pandas.read_csv(pathlib.Path.cwd() / 'results' / f"{str(y['acmi_id']).zfill(8)}.csv")
        data = data.loc[data.wikidata_id.isin([y['wikidata_id']])]
        for z in data.to_dict('records'):
            state2 = f"{z['acmi_title']} same as {z['wikidata_title']} [{z['match']}]"
            print(state2)
        print('\n')
```
